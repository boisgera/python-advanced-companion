```python
from numpy import *
from numpy.linalg import *
from matplotlib.pyplot import *
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
WEEK = 7
YEAR = 365
```

```python
N = 100
beta = 1 / (WEEK)
gamma = 1 / (2 * WEEK)
omega = 1 / YEAR
```

```python
def dSIR(t, SIR):
    S, I, R = SIR
    dS = omega * R - beta * I * S / N
    dI = beta * I * S / N - gamma * I
    dR = gamma * I - omega * R  
    return (dS, dI, dR)
```

```python
S0, I0 = 99.0, 1.0
R0 = N - S0 - I0
t_span = [0.0, 5*YEAR]
```

```python
results = solve_ivp(dSIR, t_span=t_span, y0=(S0, I0, R0))
t, y = results.t, results.y
```

```python
plt.plot(t, y.T, "-+", label=["S", "I", "R"])
plt.xlabel("Temps en jours")
plt.ylabel("Populations")
plt.legend()
pass
```

```python
t_eval = arange(0.0, 5*YEAR, 30)
```

```python
results = solve_ivp(dSIR, t_span=t_span, t_eval=t_eval, y0=(S0, I0, R0))
t, y = results.t, results.y
```

```python
plt.plot(t, y.T, "-+", label=["S", "I", "R"])
plt.xlabel("Temps en jours")
plt.ylabel("Populations")
plt.legend()
pass
```

```python
def events(t, SIR):
    return SIR[1] - 2.0
```

```python
results = solve_ivp(dSIR, y0=(S0, I0, R0), t_span=t_span, dense_output=dense_output, events=events)
tc, yc, sol, t_events = results["t"], results["y"], results["sol"], results["t_events"]
```

```python
SIR_dynamics = {
 ("S", "I"): "beta * I * S / N",
 ("I", "R"): "gamma * I",
 ("R", "S"): "omega * R"
}
```

```python
def make_fun(dynamics):
    vars = []
    for s, t in dynamics.keys():
        if s not in vars:
            vars.append(s)
        if t not in vars:
            vars.append(t)
    n = len(vars)
    
    def fun(t, state):
        ns = globals().copy()
        for var, value in zip(vars, state):
            ns[var] = value
        dstate = []
        for i in range(n):
            d = 0
            var = vars[i]
            for (edge, expr) in dynamics.items():
                source, target = edge
                if source == var:
                    d -= eval(expr, ns)
                if target == var:
                    d += eval(expr, ns)
            dstate.append(d)
        return dstate
        
    return fun 
```

```python
dSIR2 = make_fun(SIR_dynamics)
```
